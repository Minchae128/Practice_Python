```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 나눔 폰트 설치된 경로 지정 (본인 컴퓨터에 맞게 설정)
font_path = 'C:/Windows/Fonts/batang.ttc'

# 나눔 폰트의 이름 가져오기
font_name = fm.FontProperties(fname=font_path).get_name()

# 폰트 설정
plt.rc('font', family=font_name)

# 데이터 불러오기
df = pd.read_csv('학생 점수.csv')

# 결측치 확인 및 시각화
plt.figure(figsize=(10, 6))
plt.bar(df.columns, df.isnull().sum(), color='skyblue')  # 각 열의 결측치 개수를 막대 그래프로 표시
plt.xlabel('열 이름')
plt.ylabel('결측치 개수')
plt.title('열 별 결측치 개수')
plt.xticks(rotation=45)  # x축 레이블 회전하여 가독성 향상
plt.show()

# 결측치 처리
# '학생'이라는 문자열이 포함되지 않은 열을 선택하여 해당 열의 평균값으로 결측치를 채움
numeric_columns = [col for col in df.columns if '학생' not in col]
df.fillna(df[numeric_columns].mean(), inplace=True)

# 이상치 시각화 (박스 플롯)
# 모든 열을 포함한 박스 플롯을 그리고 이상치 확인
plt.figure(figsize=(12, 8))
df.boxplot()
plt.title('모든 열의 이상치 확인')  # 그래프 제목 추가
plt.xticks(rotation=45)  # x축 레이블 회전하여 가독성 향상
plt.show()

# 이상치를 제외한 박스 플롯 생성
# '학생' 열을 제외한 열들로만 구성된 박스 플롯을 그려서 이상치를 고려하지 않은 상태에서의 데이터 분포 확인
plt.figure(figsize=(12, 8))
df.drop(columns=['학생']).boxplot()
plt.title('이상치 제외한 데이터 분포')  # 그래프 제목 추가
plt.xticks(rotation=45)  # x축 레이블 회전하여 가독성 향상
plt.show()

# 결측치 =  데이터 안에 '빈값'이 있을때 결측치가 있다고 말함
# 결측치 왜 중요한가? -> 결측치가 있는 데이터를 그대로 사용하면 결과가 왜곡될 수 있음

# 이상치 = '정상적인 데이터와 너무 다른 값'을 의미 예를 들어, 
#           학급의 키를 조사하는데, 대부분의 학생들이 140cm정도 인데, 1명의 학생이 200cm라면 그 학생은 이상치
# 왜 중요한가요? 이상치가 있는 경우, 그냥 무시하면 안 되요. 
# 왜냐하면 이상치가 있는 데이터로 모델을 만들면 결과가 잘못될 수 있기 때문이에요. 이상치를 찾아내고 처리하는 것이 중요합니다.                        
```
