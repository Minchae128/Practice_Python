```python
from plotly.subplots import make_subplots
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
use_log = pd.read_csv('use_log.csv')
customer_master = pd.read_csv('customer_master.csv')
class_master = pd.read_csv('class_master.csv')
campaign_master = pd.read_csv('campaign_master.csv')
```

```python
campaign_master.loc[0, 'campaign_name'] = '일반'
campaign_master.loc[1, 'campaign_name'] = '입회비 반액 할인'
campaign_master.loc[2, 'campaign_name'] = '입회비 무료'
campaign_master
```

```python
customer_master_total = customer_master.merge(class_master, on='class').merge(campaign_master, on='campaign_id')
customer_master_total
```

```python
use_log['customer_id'] =use_log['customer_id'].astype('category')
use_log['usedate'] = pd.to_datetime(use_log['usedate'])
customer_master_total['customer_id'] =customer_master_total['customer_id'].astype('category')
customer_master_total['class'] = customer_master_total['class'].astype('category')
customer_master_total['class_name'] = customer_master_total['class_name'].astype('category')
customer_master_total['class_name'] = pd.Categorical(customer_master_total['class_name'], categories=['종일','주간','야간'])
customer_master_total['gender'] = customer_master_total['gender'].astype('category')
customer_master_total['gender'] = pd.Categorical(customer_master_total['gender'], categories=['M','F'])
customer_master_total['is_deleted'] = customer_master_total['is_deleted'].astype('category')
customer_master_total['is_deleted'] = customer_master_total['is_deleted'].replace({0:'정기', 1:'탈퇴'})
customer_master_total['campaign_id'] = customer_master_total['campaign_id'].astype('category')
customer_master_total['campaign_name'] = customer_master_total['campaign_name'].astype('category')
customer_master_total['campaign_name'] = pd.Categorical(customer_master_total['campaign_name'], categories=['일반', '입회비 반액 할인', '입회비 무료'])
customer_master_total['start_date'] = pd.to_datetime(customer_master_total['start_date'], format='%Y-%m-%d %H:%M:%S')
customer_master_total['end_date'] = pd.to_datetime(customer_master_total['end_date'], format='%Y-%m-%d %H:%M:%S')
customer_master_total['start_date(Y-m)'] = customer_master_total['start_date'].dt.strftime('%Y-%m')
customer_master_total['end_date(Y-m)'] = customer_master_total['end_date'].dt.strftime('%Y-%m')
customer_master_total['이용기간'] = customer_master_total['end_date'] - customer_master_total['start_date']
```

# 데이터 인사이트

주간반에 여성의 비율이 야간반에 남성의 비율이 높음   
스포츠센터가 주택가 주변에 있을 가능성 높음   
주간반에 여성, 야간반에 남성을 타겟으로 하여 프로그램 생성  


## 주간반 여성 회원을 위한 프로그램:
1. 모닝 요가 또는 필라테스: 스트레스 해소와 몸매 관리에 좋은 요가 또는 필라테스 클래스
2. 건강 요리 클래스: 건강한 식생활을 유도하고 간단하면서도 영양가 높은 요리를 배울 수 있는 클래스
3. 자녀 동반 프로그램: 유아 또는 어린이를 대상으로 하는 체육 활동이나 창의 놀이 클래스

## 야간반 남성 회원을 위한 프로그램:
1. 피트니스 챌린지: 체력 증진과 근육 발달을 목표로 하는 집중적인 피트니스 프로그램 (개인 또는 팀별로 경쟁을 유도)
2. 스포츠 리그: 축구, 농구, 배드민턴 등 인기 스포츠 종목을 중심으로 한 정기 리그
3. 전문가와의 워크숍: 건강, 피트니스, 금융 관리 등 남성들이 관심을 가질 만한 주제로 워크숍을 개최

종일반은 남성이 조금 많지만 비슷한 분포를 보임

## 종일반 회원을 위한 프로그램:
1. 크로스핏 클래스: 다양한 난이도와 함께 체력과 근력을 동시에 강화할 수 있는 크로스핏 클래스를 제공하여 모든 수준의 회원에게 참여를 유도
2. 그룹 스포츠: 배구, 축구, 농구 등과 같은 팀 스포츠 클래스

행사 기간내 가입자는 50% 이상이 탈퇴   

## 행사 기간내 신규 가입자를 위한 특별 관리
1. 온보딩 프로그램 신설: 신규 회원들이 체계적인 온보딩 프로그램을 통해 센터를 이용하는 방법, 각종 프로그램 소개, 건강 목표 설정 등을 배울 수 있도록 유도
2. 유연한 멤버십 옵션: 유연한 가입 옵션(월 단위 멤버십 또는 프로모션 기간 종료 후 자동 갱신되지 않는 멤버십)

1년동안 60회 이상 이용하는 회원은 정기 회원이 될 가능성 높음

## 방문을 유도 하는 이벤트 계최
1. 도전과제 프로그램: '100일 도전' 또는 '1년 중 70일 출석 도전'과 같은 프로그램을 개발하여 회원들이 정기적으로 센터를 방문하도록 동기를 부여(도전을 완료했을 때 인증서나 소정의 선물을 제공)
2. 계절별 테마 프로그램: 봄에는 야외 활동 중심의 프로그램, 여름에는 수영이나 워터 에어로빅, 가을에는 등산 준비 체력 단련 프로그램, 겨울에는 실내에서 할 수 있는 요가나 필라테스, 피트니스 특강 등 계절별로 테마를 정하여 프로그램을 계최
3. 건강 캠페인: 건강을 테마로 한 여러 워크숍, 세미나, 건강 검진 행사를 정기적으로 개최
4. 패밀리 데이: 주간반을 이용하는 주부 회원들을 대상으로 자녀와 함께 참여할 수 있는 프로그램을 주기적으로 운영
5. 리퍼럴 프로그램: 기존 회원이 새로운 회원을 데려오면 데려온 회원과 새 회원 모두에게 혜택을 주는 리퍼럴 프로그램을 도입
