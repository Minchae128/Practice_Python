만약 세 칸의 숫자가 모두 같다면, 당신은 100 마법의 별과 (같은 숫자) x 10 마법의 별을 추가로 받습니다.

```python
# 1. 당신은 오늘 친구와 만나기로 했습니다.
# 친구를 만나기 위해 운전을 해야 하는데, 출발 시간과 목적지까지 운전하는 데 필요한 시간(분 단위)를 입력받아 
# 도착 예정 시간을 계산하는 프로그램을 작성하세요.  
# 출발시간은 공백으로 구분할 수 있습니다. 시간은 24시간 단위입니다.(+ 오전오후로 돌아가는 12시간으로도 해보기 단, 다풀었을때.)
start_time = input('출발 시간을 입력하시오 (시간 분): ')
load_time = int(input('출발 시간과 목적지까지 운전하는 데 필요한 시간(분 단위)을 입력하시오: '))

start_hour, start_minute = start_time.split()

start_hour = int(start_hour)
start_minute = int(start_minute)

# 출발 시간과 운전 시간을 더한 후, 시간과 분을 계산합니다.
end_hour = start_hour + (start_minute + load_time) // 60
end_minute = (start_minute + load_time) % 60

# 만약 도착 분이 10 이하면, 0을 추가해줍니다.
if end_minute < 10:
    arrival_time = f"{end_hour}시 0{end_minute}분"
else:
    arrival_time = f"{end_hour}시 {end_minute}분"

print(f'도착 예정 시간은: {arrival_time} 입니다.')
```

```python
# 2. for 반복문을 통해 구구단을 2단부터 9단까지 출력해 주는 프로그램을 작성하세요

for i in range(1, 10):
 for j in range(1, 10):
    print(f'{i} * {j} = {i * j}')
```

```python
# 3. 10개의 서로 다른 자연수를 입력했을 때 최댓값과 위치 인덱스를 출력하세요.
list_num = []

for i in range(10):
    num = int(input("자연수를 입력하세요: "))
    list_num.append(num)

max_value = list_num[0]  
max_index = 0

for i in range(1, len(list_num)):
    if list_num[i] > max_value:
        max_value = list_num[i]
        max_index = i

print('최대 값은:', max_value)
print('최대 값의 위치 인덱스는:', max_index)

```

```python
# 4. 입력한 영단어에서 가장 많이 사용된 알파벳을 찾고 대문자로 출력하는 프로그램을 작성하세요 
# 가장 많이 사용된 알파벳이 2개 이상인 경우 ???를 출력하세요(리스트/ 집합)
eng = input("영단어를 입력하시오: ")

word = set(eng.upper())

list_count = []

for i in word:
    count = eng.count(i)
    list_count.append((i, count))

max_count = 0
most_A = []

for i, count in list_count:
    if count > max_count:
        max_count = count
        most_A = [i]
    elif count == max_count:
        most_A.append(i)

if len(most_A) == 1:
    print("가장 많이 사용된 알파벳은", most_A[0], "입니다.")
else:
    print("가장 많이 사용된 알파벳이 2개 이상입니다. ???를 출력합니다.")
```

```python
# 5. 당신은 시험 점수를 관리하는 교사입니다.  
# 최근 실시한 시험에서 학생들이 받은 점수는 다음과 같습니다 exam_scores = [95, 68, 85, 71, 23, 75, 91, 54, 35, 48]  
# 학생들이 추가로 제출한 과제를 평가한 후 각 학생의 시험 점수에 다음과 같이 추가 점수를 부여하기로 결정했습니다.
# extra_points = [1, 10, 8, 9, 8, 5, 6, 4, 7, 0]  
# 각 학생들의 번호, 시험점수, 과제점수, 최종점수를 출력하고 
# 최종점수가 80점이상인 학생들의 번호를 "80점 이상인 학생 번호:" 라는 메세지와 함께 출력하세요
exam_scores = [95, 68, 85, 71, 23, 75, 91, 54, 35, 48]
extra_points = [1, 10, 8, 9, 8, 5, 6, 4, 7, 0]  

print("학생번호  시험점수  과제점수  최종점수")
print("----------------------------------")

for i in range(len(exam_scores)):
     sum = exam_scores[i] + extra_points[i]
      
     if sum >= 80:
        print(f"80점 이상인 학생 번호: {i + 1}")  
        print(f"{i + 1}           {exam_scores[i]}       {extra_points[i]}    {sum}")

     else:
        print(f"{i + 1}           {exam_scores[i]}       {extra_points[i]}    {sum}")                      
```
