```python
# 1
import pandas as pd

df = pd.read_excel('CARD_SUBWAY_MONTH_202403.xlsx') 
df

df.head(5)
```

```python
# 2
df.info()

df.isnull()
```

```python
# 3
df.info()
```

```python
# 3
df['사용일자'] = pd.to_datetime(df['사용일자'])

df.info()
```

```python
# 4
df.groupby(['노선명', '역명'])[['승차총승객수', '하차총승객수']].agg({'승차총승객수': ['mean', 'max', 'min'], '하차총승객수': ['mean', 'max', 'min']})
```

```python
# 5
df_total = df.groupby(['노선명', '역명']).filter(lambda x: (5000 <= x['승차총승객수']).all())[['노선명', '역명', '승차총승객수']]

df_total
```

```python
# 6
df['사용일자'] = pd.to_datetime(df['사용일자'])
df['요일'] = df['사용일자'].dt.day_name()

df
```

```python
# 7
df.sort_values('승차총승객수', ascending=False)
```

```python
# 8
# 0~6번 인덱스 위치에 각 한글 요일 이름이 담긴 리스트
weekday_list = ['월', '화', '수', '목', '금', '토', '일']

# 사용일자 열의 값(요일)을 한글 요일로 변환
df['요일'] = df['사용일자'].apply(lambda x: weekday_list[x.weekday()])

df
```

```python
# 9 
df.groupby('요일').agg({'승차총승객수': ['mean', 'sum'], '하차총승객수': ['mean', 'sum']})
```

```python
#10 
#df.groupby('역명').agg({'승차총승객수': 'sum', '하차총승객수': 'sum'}).sum(axis=1).sort_values(ascending=False).head(10)


#top_10 = df.groupby('역명').agg({'승차총승객수': 'sum', '하차총승객수': 'sum'}).sum(axis=1).nlargest(10)

top_10 = pd.DataFrame(df.groupby('역명').agg({'승차총승객수': 'sum', '하차총승객수': 'sum'}).sum(axis=1).sort_values(ascending=False).head(10))
top_10.columns = ['Top10']
top_10
```
